# Description


This file will call the procedures, views and triggers which are created in MySQL Workbench. Initially we import all the necessary libraries and then create connection between python and MySQL Workbench. Then we call all the procedures which we have built in MySQL Workbench for the below mentioned 4 tasks. The input for performing any task is taken from the user. For example if the user wants to add a review then the code will take input from the user regarding the reaturant name, review rating, comment etc.Then the task will be executed and the output will be displayed to the user.

Task 1 : Review a Restaurant

Task 2 : Reserve a table at Restaurant

Task 3 : Cancel Reservation

```python
import pandas as pd
from datetime import datetime
from datetime import timedelta
from bs4 import BeautifulSoup
from tabulate import tabulate
```

```python
import pymysql
#connection between python and mysql workbench
conn=pymysql.connect(host='localhost',port=int(3306),user='root',passwd='root',db='r3_db_system')
cursor= conn.cursor()
```

```python
# Function to add a review to a restaurant
def review_restaurant():
    print("Review a Restaurant")
    restaurantName=input("Enter restaurant name : ")
    userName=input("Enter user first name : ")
    reviewComment=input("Enter review comment : ")
    likedOrNot=int(input("Did you like the restaurant? (yes:1,no:0) : "))
    userRating=input("Enter user rating : ")
    print("\n --- OUTPUT---")
#     reviewComment="Amazing experieance"
#     likedOrNot=1
#     userRating=4.7
#     restaurantName='Shake Shack'
#     userName='Lenna'
    cursor.execute("call add_restaurant_review(%s, %s, %s, %s,%s);", (reviewComment, likedOrNot, userRating, restaurantName, userName))
    cursor.execute("select * from restaurant_review order by review_id desc limit 1;")
    records=cursor.fetchall()
    print(' \nReview Details:')
    for row in records:
        print(row, '\n')
    conn.commit()
    print("\nEnd Task")
    
# review_restaurant()
```

```python
# Function to reserve a restaurant
def reserve_restaurant():
    print('Reserve a table at a restaurant')
    restaurantName=input("Enter restaurant name : ")
    userName=input("Enter user first name : ")
    noOfPeople=int(input("Enter no of people : "))
    reservationDate=input("Enter Date (YYYY-MM-dd) : ")
    reservationTime=input("Enter Time (HH:MM:SS) : ")
    print("\n --- OUTPUT---")
#     noOfPeople=4
#     reservationDate='2022-06-14'
#     reservationDate = datetime.strptime(reservationDate, '%Y-%M-%d').date()
#     reservationTime='21:15:25'
#     restaurantName='Subway'
#     userName='Sage'
    
    cursor.execute("SELECT * from restaurant where restaurant_name=%s;",(restaurantName))
    records=cursor.fetchall()
    print(' \nRestaurant Details Before reservation')
    for row in records:
        print(row, '\n')
    conn.commit()
    
    cursor.execute("CALL restaurant_reservation (%s,%s,%s,%s,%s);", (noOfPeople, reservationTime, reservationDate, restaurantName, userName))
    cursor.execute("SELECT max(reservation_id), no_of_people, reservation_time, reservation_date, restaurant_id, user_id from restaurant_reservation")
    records=cursor.fetchall()
    print('\nReservation Deatils')
    for row in records:
        print(row, '\n')
    conn.commit()
    
    cursor.execute("SELECT * from restaurant where restaurant_name=%s;",(restaurantName))
    records=cursor.fetchall()
    print('\nRestaurant Details After reservation')
    for row in records:
        print(row, '\n')
    conn.commit()
    
    print("\nEnd Task")
    
# reserve_restaurant()
```

```python
# Function to cancel a reservation at a restaurant
def cancel_reservation():
    print('\nCancel a Reservation')
    cancellationReason=input("Enter cancellation reason : ")
    restaurantName=input("Enter restaurant name : ")
    reservationId=input("Enter reservationId : ")
    print("\n --- OUTPUT---")
#     cancellationReason="Did not like"
#     restaurantName='Dominos'
#     reservationId=9002

    cursor.execute("SELECT * from restaurant where restaurant_name=%s;",(restaurantName))
    records=cursor.fetchall()
    print('\nRestaurant Details Before cancellation')
    for row in records:
        print(row, '\n')
    conn.commit()
    
    cursor.execute("call restaurant_cancellation(%s, %s);", (cancellationReason, reservationId))
    cursor.execute("select * from restaurant_cancellation order by cancellation_id desc limit 1;")
    records=cursor.fetchall()
    print('\nCancellation Details')
    for row in records:
        print(row, '\n')
    conn.commit()
    
    print('\nRestaurant Details After cancellation')
    cursor.execute("SELECT * from restaurant where restaurant_name=%s;",(restaurantName))
    records=cursor.fetchall()
    for row in records:
        print(row, '\n')
    conn.commit()
    print("\nEnd Task")
    
# cancel_reservation()
```

```python
continue_option='Y'
while continue_option=='Y' or continue_option=='y':
    print('\nSelect Among the Following Tasks')
    print('Task 1 : Review a Restaurant')
    print('Task 2 : Reserve a table at Restaurant')
    print('Task 3 : Cancel Reservation')
#     print('\nTask 4 : Search for Restaurants')
    choice=int(input("\nEnter your choice : "))
    if choice==1:
        review_restaurant()
    elif choice==2:
        reserve_restaurant()
    elif choice==3:
        cancel_reservation()
    else:
        print('\ninvalid choice\n')
#     elif choice==4:
#         search_restaurant()
    continue_option=input('\nWould you like to continue ? (yes:y/Y, no: n/N) : ')

```
